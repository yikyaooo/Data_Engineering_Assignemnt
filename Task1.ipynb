```python
from classes.NewsProducer import NewsProducer
from classes.NewsConsumer import NewsConsumer
```

```python
# Kafka Configuration
KAFKA_BROKER = 'localhost:9092'  # Change if using a different setup
KAFKA_TOPIC = 'nation_news'
HDFS_URL = "http://localhost:9000"  # Change to your HDFS Namenode URL
LOCAL_PATH = "/tmp/news1.json" 
HDFS_PATH = "/user/student/raw_data/news1.json"  # Directory in HDFS
```

```python
# Generate URLs for the last 20 days
from datetime import datetime, timedelta

base_url = "https://www.thestar.com.my/news/nation?sdate={}&edate={}"
num_days = 20

# Get today's date
today = datetime(2020, 12, 1)

# Create a list of URLs for the last 20 days
links = [
    base_url.format((today - timedelta(days=i)).strftime("%Y-%m-%d"),
                    (today - timedelta(days=i)).strftime("%Y-%m-%d"))
    for i in range(num_days)
]
```

```python
producer = NewsProducer(KAFKA_BROKER, KAFKA_TOPIC)

for link in links:
    producer.fetch_and_produce(link)

producer.close()
```

```python
consumer = NewsConsumer(KAFKA_BROKER, KAFKA_TOPIC, "natiob_groups", LOCAL_PATH, HDFS_PATH)

```

```python
articles = consumer.consume_articles()
```

```python
if articles:
    consumer.save_to_hdfs(articles)
else:
    print("No article to save")
```
